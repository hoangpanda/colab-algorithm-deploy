```python
import sympy as smp
import math
import numpy as np
import matplotlib.pyplot as plt

def fact(n):
    return math.factorial(n)

def gamma(z):
    x = smp.Symbol('x')
    return smp.integrate(x**(z-1)*smp.exp(-x), (x, 0, 999))

def f(alpha, beta, lb):
    return gamma(alpha+beta)/(gamma(alpha)*gamma(beta))*lb**(alpha-1)*(1-lb)**(beta-1)


x = np.linspace(0, 1, 100)
y = [f(4, 12, xx) for xx in x]

#print(y)

plt.plot(x, y, 'g')
plt.show()


```
